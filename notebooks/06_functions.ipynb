## 🔥 Scenario 1: Simple Function
Write a function called `classify_number(x)` that:
1. Returns `"Negative"` if x < 0  
2. Returns `"Zero"` if x == 0  
3. Returns `"Positive"` if x > 0  
### ⚠️ Requirements
- Use a function (NOT just print)
- Use `return`, not `print` inside the function

```python
def classify_number(x):
    if x < 0:
        return "Negative"
    elif x == 0:
        return "Zero"
    else:
        return "Positive"

print(classify_number(3))
```

## 🔥 Scenario 2: Function with List Processing
Write a function called `sum_even(nums)` that:
1. Takes a list of numbers
2. Returns the sum of ONLY even numbers
### ⚠️ Requirements
- Use a loop inside the function
- Do NOT use list comprehension or sum()
- Use `return`

```python
def sum_even(nums):
    total = 0
    for number in nums:
        if number % 2 == 0:
            total += number
    return total

nums = [3, 10, 5, 18, 7, 20]
print(sum_even(nums))
```

## 🔥 Scenario 3: Function with Conditional Filtering
Write a function `filter_numbers(nums, threshold)` that:
1. Takes:
   - a list of numbers
   - a threshold value
2. Returns a list of numbers that are:
   - GREATER than threshold
   - AND EVEN
### ⚠️ Requirements
- Use a loop (NOT list comprehension)
- Use `return`
- Do NOT modify original list

```python
def filter_numbers(nums, threshold):
    result = []
    for number in nums:
        if number > threshold and number % 2 == 0:
            result.append(number)
    return result

nums = [5, 12, 7, 18, 3, 20, 9]
print(filter_numbers(nums, 10))
```

## 🔥 Scenario 4: Function with Multiple Results
Write a function `analyze_numbers(nums)` that:
1. Returns TWO values:
   - sum of even numbers
   - count of odd numbers
### ⚠️ Requirements
- Use a loop
- Use `return`
- Return BOTH values (not print)

```python
def analyze_numbers(nums):
    even_sum = 0
    odd_count = 0

    for num in nums:
        if num % 2 == 0:
            even_sum += num
        else:
            odd_count += 1

    return even_sum, odd_count

nums = [3, 10, 5, 18, 7, 20]
even_sum, odd_count = analyze_numbers(nums)

print(even_sum)   # 48
print(odd_count)  # 3
```

## 🔥 Scenario 1: While Loop Counting
Write a program that:
1. Uses a `while` loop
2. Prints numbers from 1 to 5
### ⚠️ Requirements
- Use a `while` loop (NOT a for loop)
- Manually update the counter

```python
x = 0

while x < 5:
    x += 1
    print(x)
```
