# Lists and Randomization

```python
import random
```

```python
# random integers
random_integer = random.randint(2, 7)
print(random_integer)
```

```python
# random floating point between 0 and 1 (0 included, 1 is not)
random_number = random.random() * 10
print(random_number)
```

```python
# random floating point between a and b inclusive
random_float = random.uniform(1, 7)
print(random_float)
```

### Print Random Heads and Tails

```python
random_heads_or_tails = random.randint(0,1)
if random_heads_or_tails == 0:
    print("Heads")
else:
    print("Tails")
```

### Working with Lists

```python
fruits = ["orange", "banana", "grapes", "blueberry"]
print(fruits[2])
print(fruits[-1])
```

```python
# change blueberry to strawberry
fruits[-1] = "strawberry"
print(fruits)
```

```python
# add a single item to the list
fruits.append("blackberry")
```

```python
print(fruits)
```

```python
# extending a list
fruits.extend(["mango", "pear"])
print(fruits)
```

### Nested Lists

```python
fruits = ["orange", "banana", "grapes", "blueberry"]
vegetables = ["spinach", "kale", "tomatoes", "celery", "potatoes"]

dirty_dozen = [fruits, vegetables]

print(dirty_dozen)
```

```python
print(dirty_dozen[1][1])
```

```python
print("What do you choose?")
choice = int(input("Type 0 for Rock, 1 for Paper or 2 for Scissors"))

print(f"You chose {choice}")

import random
computer_choice = random.randint(0,2)
print(f"The computer chose {computer_choice}")

if choice == computer_choice:
     print("It's a draw")
elif choice == 0:
    if computer_choice == 1:
        print("You lose")
    else:
        print("You win")
elif choice == 1:
    if computer_choice == 0:
        print("You win")
    else:
        print("You lose")
elif choice == 2:
    if computer_choice == 0:
        print("You lose")
    else:
        print("You win")
else:
    print("You typed an invalid number")
```
