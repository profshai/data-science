## 🔥 Scenario 1: Number Classification

You are given a variable `x`.

Write a program that:

1. Prints `"Invalid"` if `x` is not a number  
2. Prints `"Negative"` if `x < 0`  
3. Prints `"Zero"` if `x == 0`  
4. Prints `"Small Positive"` if `0 < x < 10`  
5. Prints `"Large Positive"` if `x >= 10`  

---

### ⚠️ Requirements
- Use `if / elif / else`
- Do **not** repeat conditions unnecessarily
- Handle non-numeric inputs safely (no errors)


```python
try:
    x = float(input("Type a number: "))
    
    if x < 0:
        print("Negative")
    elif x == 0:
        print("Zero")
    elif 0 < x < 10:
        print("Small Positive")
    else:
        print("Large Positive")
except ValueError:
    print("Invalid")
```

## 🔥 Scenario 2: Exam Grade Classification
You are given a variable `score`.
Write a program that:

1. Prints `"Invalid"` if `score` is not a number
2. Prints `"Invalid"` if `score` is less than 0 or greater than 100
3. Prints `"A"` if `score` is between 90 and 100
4. Prints `"B"` if `score` is between 80 and 89
5. Prints `"C"` if `score` is between 70 and 79
6. Prints `"D"` if `score` is between 60 and 69
7. Prints `"F"` if `score` is below 60

### ⚠️ Requirements
- Use `if / elif / else`
- Handle invalid input safely
- Do not repeat conditions unnecessarily

```python
try:
    score = float(input("What is your score?"))
    print(f"Your score is {score}")

    if 0 <= score < 60:
        print("F")
    elif score < 70:
        print("D")
    elif score < 80:
        print("C")
    elif score < 90:
        print("B")
    elif score <= 100:
        print("A")
    else:
        print("Invalid")
   
except ValueError:
    print("Invalid")
```

## 🔥 Scenario 3: Login System
You are given:
- `username`
- `password`

Valid credentials:
- username = "admin"
- password = "1234"

### Write a program that:
1. Prints `"Invalid"` if either input is empty  
2. Prints `"Username not found"` if username is incorrect  
3. Prints `"Wrong password"` if username is correct but password is wrong  
4. Prints `"Login successful"` if both are correct  

```python
username = input("What is your username? ")
password = input("What is your password? ")

# Step 1: Validate input
if username == "" or password == "":
    print("Invalid")

# Step 2: Check username
elif username != "admin":
    print("Username not found")

# Step 3: Check password
elif password != "1234":
    print("Wrong password")

# Step 4: Success
else:
    print("Login successful")
```

## 🔥 Scenario 4: Discount System
You are given:
- `age`
- `is_member` (True or False)

### Rules:
1. Print `"Invalid"` if age is not a number or age < 0  
2. If age < 18:
   - If member → `"10% discount"`
   - Else → `"5% discount"`
3. If age >= 18:
   - If member → `"20% discount"`
   - Else → `"No discount"`

### ⚠️ Requirements
- Use nested `if` statements
- Handle invalid input safely
- Avoid repeating conditions

```python
try:
    age = float(input("What is your age?"))
    is_member = int(input("Are you a member? Enter 1 for yes or 0 for no"))
    if age < 0:
        print("Invalid")
    elif age >= 18:
        if is_member == 1:
            print("20% discount")
        else:
            print("No discount")
    elif age > 0:
        if is_member == 1:
            print("10% discount")
        else:
            print("5% discount")
    
   
except ValueError:
    print("Invalid")
```
