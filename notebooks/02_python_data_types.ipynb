### BMI Calculator

```python
height = 1.65
weight = 84
bmi = weight / (height)**2
print(bmi)
print(int(bmi)) #flooring
print(round(bmi)) #rounding to whole number
print(round(bmi, 2)) #rounding to the nearest decimal places
```

## Day 2 Project

```python
print("Welcome to the tip calculator")
total_bill = float(input("What was the total bill?"))
tip = int(input("How much tip would you like to give? 10, 12, or 15"))
tip_amount = total_bill * tip/100
number_of_people = int(input("How many people to split the bill?"))
each_amount = round((total_bill + tip_amount) / number_of_people, 2)

print(f"Each person should pay: {each_amount}")
```
